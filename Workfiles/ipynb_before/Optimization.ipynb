```python
import numpy as np
import matplotlib.pyplot as plt

def sine(x):
    return np.sin(x)

x = np.arange(-10, 10, 0.1)
plt.plot(x, sine(x))

from scipy import optimize

# The default (Nelder Mead)
result = optimize.minimize(sine,x0=3, method='Nelder-Mead', bounds=None ,options={'disp': True, 'return_all': True,})
xf_sine = result["x"]
plt.plot(xf_sine, sine(xf_sine), "o")

#BFGS
result2 = optimize.minimize(sine, x0=3, method='BFGS', jac=None, tol=0.001 , options={'gtol': 1e-05, 'norm': np.inf, 'eps': 1.4901161193847656e-08, 'maxiter': None, 'disp': True, 'return_all': True, 'finite_diff_rel_step': None})
xf2_sine = result2["x"]
plt.plot(xf2_sine, sine(xf2_sine), "o")

jac_sine = result2["jac"]
plt.plot(jac_sine, "o")

#CG
result3 = optimize.minimize(sine, x0=3, method='CG', jac=None, tol=0.001, options={'gtol': 1e-05, 'norm': np.inf, 'eps': 1.4901161193847656e-08, 'maxiter': None, 'disp': True, 'return_all': True, 'finite_diff_rel_step': None})
xf3_sine = result3["x"]
plt.plot(xf3_sine, sine(xf3_sine), "o")
jac_sine2 = result3["jac"]
plt.plot(jac_sine2, "o")

#Newton-CG
result4 = optimize.minimize(sine, x0 = 2,method='Newton-CG', jac= sine, hess=sine, hessp=None, tol=0.001, options={'xtol': 1e-05, 'eps': 1.4901161193847656e-08, 'maxiter': None, 'disp': True, 'return_all': True})
xf4_sine = result4["x"]
plt.plot(xf4_sine, sine(xf4_sine), "o")
jac_sine3 = result4["jac"]
plt.plot(jac_sine3, "o")
```
